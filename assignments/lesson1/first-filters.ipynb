```python
import pandas as pd
import  datetime
path="../../../lesson_1/people.csv"
pd.set_option('display.max_colwidth', None)
df = pd.read_csv(path)
df.tail(10)
```

```python
# filter by birthplace Grudziądz and male and age between 19 and 40
# df = df[df['birthplace'] == 'Grudziądz']
# df = df[df['gender'] == 'male']
df = df[df['birthPlace'] == 'Grudziądz']
df = df[df['gender'] == 'M']
# add column with age : calcuale age based on current year 2026 and birthDate column which is in format YYYY-MM-DD
df['birthDate'] = pd.to_datetime(df['birthDate'])
df['age'] = 2026 - df['birthDate'].dt.year
# create `id` column equal to the current DataFrame index
df = df.copy()
df["id"] = df.index

df.head(2)


```

#### extarct only id & job  and convert to json 


```python
only_job_df = df[["id", "job"]].copy()


only_job_df

# Start Generation Here
import json

# convert DataFrame to list of {"id": ..., "job": ...}
jobs_records = only_job_df.to_dict(orient="records")

# pretty JSON (supports Polish characters)
jobs_json = json.dumps(jobs_records, ensure_ascii=False, indent=2)

print(jobs_json)

# optional: save to file
with open("filtered_jobs.json", "w", encoding="utf-8") as f:
    f.write(jobs_json)
# End Generation Her

```
