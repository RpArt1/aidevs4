```python
import tiktoken

def count_tokens(text: str, model: str = "gpt-4o") -> int:
    enc = tiktoken.encoding_for_model(model)
    return len(enc.encode(text))
```

# simple call to LLM 

```python
import os
import sys
sys.path.insert(0, os.path.abspath(".."))

from common.llm_service import LLMService



llm = LLMService()

prompt = "Hello, can you summarize the quick brown fox and lazy dog story?"
model_name = "google/gemini-3.1-flash-lite-preview"

llm.model = model_name
print(f"Sending prompt to {model_name}")
response = llm.chat(
    messages=[{"role": "user", "content": prompt}],
    temperature=0.2
)

print(f"Response received ({len(response)} chars)")
print("Model response:", response)

input_tokens = count_tokens(prompt)
output_tokens = count_tokens(response)
print(f"\nInput tokens:  {input_tokens}")
print(f"Output tokens: {output_tokens}")
```

# call with structured output 


```python
import json

CLASSIFICATION_SCHEMA = {
    "type": "json_schema",
    "name": "classification",
    "strict": True,
    "schema": {
        "type": "object",
        "properties": {
            "sentiment": {
                "type": "string",
                "enum": ["positive", "negative", "neutral"],
                "description": "The sentiment of the text."
            },
            "confidence": {
                "type": "number",
                "description": "Confidence score from 0 to 1."
            }
        },
        "required": ["sentiment", "confidence"],
        "additionalProperties": False
    }
}

text = "I absolutely love this new phone, the camera is incredible!"

messages = [
    {"role": "system", "content": "Classify the sentiment of the given text."},
    {"role": "user", "content": text},
]

response = llm.chat_structured(messages, schema=CLASSIFICATION_SCHEMA)

print(f"Result: {response}")
```
